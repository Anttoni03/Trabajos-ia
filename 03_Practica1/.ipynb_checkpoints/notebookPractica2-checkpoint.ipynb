## Introducció C

Per a l'entrenament dels models de classificació, primer s'obtenen les dades, usant `load_digits()`.

Aquests, que contenen les imatges (propietats del model) i els resultats, es separen en dades d'entrenament (`X_train, y_train`) i de proves (`X_test, y_test`) amb un 50%, de forma que, com no és un *dataset* molt gran, pugui tenir una considerable quantitat de test.

A més, es mostra un exemple de 10 de les imatges per utilitzar, una de cada classe amb la seva respectiva etiqueta amunt.

```python
# Colors útils per a tot el notebook
col = {
    'main': '#cd1f36',
    'mainD': '#723640',
    'dark': '#26373D',
    'second': '#388BAA',
    'third': '#C79B27',
    'bright': '#FDF0DC'
}
```

```python
# Funció per facilitar l'impressió d'imatges de prediccions
def mostrarPredicciones(X_t, y_t, pred):
    _, axes = plt.subplots(nrows=2, ncols=6, figsize=(9, 4))
    plt.tight_layout(h_pad=1)
    for ax, image, labelPred, labelReal in zip(axes.ravel(), X_t, pred, y_t):
        ax.set_axis_off()
        image = image.reshape(8, 8)
        rect = patches.Rectangle((-0.5,-0.5), image.shape[1], image.shape[0], linewidth=3, edgecolor=col['main'], facecolor='none')
        ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
        ax.add_patch(rect)
        ax.set_title(f"Real: {labelReal}\nPredicció: {labelPred}")

# Funció per facilitar l'impressió d'informes de predicció
def showReport(X_t, y_t, pred, frase):
    mostrarPredicciones(X_t, y_t, pred)
    disp = metrics.ConfusionMatrixDisplay.from_predictions(y_t, pred, cmap='Reds')
    disp.figure_.suptitle(("Matriu de confusió de: " + frase))
    plt.show()
    print(classification_report(y_t, pred))
    return classification_report(y_t, pred, output_dict=True)
```

```python
# Es carrega el dataset de digits bàsic
digits = load_digits()

# Aquí es divideixen les dades en 50% entrenament - 50% proves
X_train, X_test, y_train, y_test = train_test_split(
    digits.data, digits.target, test_size=0.5, shuffle=False
)
```

```python
_, axes = plt.subplots(nrows=1, ncols=10, figsize=(17, 4))
plt.tight_layout()
for ax, image, label in zip(axes, digits.images, digits.target):
    ax.set_axis_off()
    rect = patches.Rectangle((-0.5,-0.5), image.shape[1], image.shape[0], linewidth=3, edgecolor=col['main'], facecolor='none')
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    ax.add_patch(rect)
    ax.set_title("Nombre: %i" % (label))
```

## Predicció de Regressió logística

Pel model de regressió logística, s'usa `LogisticRegression()`, de forma que retorna un objecte de regressió logística apte pel problema. Després, es fan servir les dades d'entrenament amb la funció **fit** i es genera realment el model d'aprenentatge per aquest problema en concret. Més endavant amb la funció **predict** s'obté la seva predicció de les dades de proves i genera una mostra de 12 prediccions. També acaba generant un informe.

Aquest tipus d'informe és el que principalment es veurà durant tot el notebook i compta amb la matriu de confusió de les prediccions aconseguides referents als resultats reals i a un informe de la classificació feta. Aquest compta amb càlculs fets automàticament com *precision*, *recall* o *f1-score*

<p class="nota"> Nota: Els models es generen sense paràmetres en aquest apartat, però al model de regressió logística sorgeix un error amb el paràmetre per defecte <em>max_iter</em>, que demana augmentar el seu valor per poder arribar al final de l'entrenament.</p>

```python
logisticReg = LogisticRegression(max_iter = 5000)
logisticReg.fit(X_train, y_train)
logRegPred = logisticReg.predict(X_test)
logRegReport = showReport(X_test, y_test, logRegPred, "Regressió logística")
```

Aquí es pot veure que és un bon model, ja que s'han reconegut quasi tots els nombres de forma correcta. Visualment, es veu a la matriu que la classe 1 la 3 es confonen i es classifiquen algunes vegades com 9 i 8 respectivament, però sense molta més equivocació notòria.

Al càlcul de precisió, es veu que les dades són molt bones, amb un 93% de **accuracy**, i altres valors que no es fan molt enfora. Aquest és el cas de les mitges no ponderades de cada classe:

| Precision | Recall | F1-Score |
| --------- | ------ | --------- |
|    94%    |   93%  |    93%   |

## K-fold Crossvalidation

El mètode de *K-fold Crossvalidation* s'aplica als models d'aprenentatge, dividint les dades d'entrenament en k seccions i avaluant k vegades aquestes divisions, variant els paràmetres, per saber quin funciona millor per les dades, de forma que aquest retorna els millors hiperparàmetres a usar.

<img class="border" src="cv.png" align="right" width="300"/>

<p> En aquest cas, la k serà 5, de forma que es farà un <strong>5-fold Crossvalidation</strong>. La forma escollida de fer aquesta validació és amb <code>GridSearchCV()</code>, que li pots indicar una sèrie de paràmetres i opcions dins cada paràmetre i, entrenant aquest objecte amb el model i les dades d'entrenament, retorna dades útils com  <code>best_params_</code> (els paràmetres amb millors resultats).</p>

<p> La forma d'assignar la k és amb <code>cv=5</code>, encara que es podria haver creat un objecte <code>KFold(n_splits=5)</code> i assignar-ho a l'objecte <i>GridSearchCV</i>. Més endavant s'entrena aquesta cerca i s'obtenen els millors paràmetres, aquests a una variable per a cada model.</p>

<br>

L'explicació per a cada conjunt de paràmetres és la següent:

- **Regressió logística**
  - <b>C</b> : Aquesta variable és la inversa de la força que regularitza l'aprenentatge. Per aquest motiu, del valor per defecte (1), s'han afegit valors més baixos.
  - <b>solver</b> : Aquesta variable representa l'algorisme d'optimització del problema. Els algorismes escollits per aquest problema són el valor per defecte (*lbfgs*), *newton-cg* i *liblinear*. Això perquè aquests dos darrers són molt bons per conjunts de dades petits i mitjans.
- **Perceptró**
  - <b>alpha</b> : Aquesta variable indica la regularització dels coeficients del model. Els valors més petits ajuden a l'entrenament, però generen sobre ajustament. A partir del valor per defecte (0.0001), he afegit valors més baixos i més alts
- **Arbre de decisió**
  - <b>max_depth</b> : Aquesta és la profunditat màxima a la qual arribarà la construcció de l'arbre. S'han triat diversos valors arbitràriament, ja que el valor per defecte depèn del nombre de *min_samples_split*
  - <b>min_samples_split</b> : Aquest és el nombre mínim de mostres per dividir un node intern de l'arbre. Això vol dir que com menor el valor, més fàcilment hi haurà més nodes. Com el valor per defecte és 2 (el mínim disponible), hem afegit valors propers.
- **Bosc aleatori**
  - <b>n_estimators</b> : Aquest és el nombre d'arbres dins el bosc. Per aquest motiu, té relació directa amb el cost i rendiment del model. Com el valor per defecte és 100, hem afegit la meitat i el doble.
  - <b>max_depth</b> : Aquesta és la profunditat màxima a la qual arribarà la construcció de cada arbre. S'han triat diversos valors arbitràriament, més petits que l'Arbre de decisió, ja que el valor per defecte depèn del nombre de *min_samples_split* i és un model més costós.
  - <b>min_samples_split</b> : Aquest és el nombre mínim de mostres per dividir un node intern de cada arbre. Això vol dir que com menor el valor, més fàcilment hi haurà més nodes. Com el valor per defecte és 2 (el mínim disponible), hem afegit un valor proper.

```python
# definir els diccionaris de paràmetres per a cada model
param_grid_rl = {'C': [0.01, 0.1, 1], 'solver': ['newton-cg','lbfgs','liblinear']}
param_grid_p = {'alpha': [1e-9, 1e-6, 1e-3, 0.1, 1, 10]}
param_grid_ad = {'max_depth': [10,20,30,40] , 'min_samples_split': [2,3,4]}
param_grid_ba = {'n_estimators': [50, 100, 200], 'max_depth': [5,10], 'min_samples_split': [2,4]}

# es creen tants GridSearchCV com models, i s'assignen
grid_search_rl = GridSearchCV(logisticReg, param_grid_rl, cv=5)
grid_search_p = GridSearchCV(perceptron, param_grid_p, cv=5)
grid_search_ad = GridSearchCV(decisTree, param_grid_ad, cv=5)
grid_search_ba = GridSearchCV(randomForest, param_grid_ba, cv=5)

# s'ajusten les dades per a cada model
grid_search_rl.fit(X_trainB, y_trainB)
grid_search_p.fit(X_trainB, y_trainB)
grid_search_ad.fit(X_trainB, y_trainB)
grid_search_ba.fit(X_trainB, y_trainB)

# s'btenen els millors paràmetres
rlParams = grid_search_rl.best_params_
pParams = grid_search_p.best_params_
adParams = grid_search_ad.best_params_
baParams = grid_search_ba.best_params_
```

Després, es creen nous models (aquesta vegada amb notació amb una B com a sufix, per ser l'apartat B). Aquests són creats amb els paràmetres obtinguts com a “millors” i entrenats amb les dades d'entrenament de l'apartat B.

```python
print("Millors paràmetres de Regressió logística:\t", rlParams)
print("Millors paràmetres de Perceptró:\t\t", pParams)
print("Millors paràmetres d'Arbre de decisió:\t\t", adParams)
print("Millors paràmetres de Bosc aleatori:\t\t", baParams)
```

## Introducció A

Les tres característiques escollides són:

- **Quantitat de píxels negres** : Es compta la quantitat de píxels negres de la imatge. Com cada píxel està representat per un enter de 0 (blanc) a 16 (negre), s'ha considerat un píxel com a negre si el seu valor és superior a 8.
- **Quantitat de cercles** : És un comptador de cercles de la imatge. Aquest comptador no és infal·lible, i pot fallar, però el seu marge d'error és acceptable per veure si la característica és vàlida o no. Consisteix a convertir cada imatge en píxels negres i blancs totals (valor superiors a 7 seran píxels negres) per decidir si els possibles cercles dels números són cercles tancats o oberts. Després, va cercant llocs a les imatges on els costats laterals i superiors siguin tancats. A partir d'aquí es comprova la vertical inferior. Si aquesta és tancada i els seus contigus laterals continuen essent-ho, es considera que s'ha trobat un cercle. Aquesta característica va dirigida, per exemple, a millorar el reconeixement de la classe 8, ja que és l'única que pot contenir 2 cercles.
- **Densitat en els quadrants** : Classifica quin dels 4 quadrants equivalents de cada imatge té major densitat de píxels. El que fa és sumar cada valor de píxel de cada un dels 4 quadrants principals (superior esquerra, superior dreta, inferior esquerra, inferior dreta) i comparar. Assigna un valor a cada quadrant.

## Característiques per computador

### Quantitat de píxels negres

Aquí es compta la quantitat de píxels amb valor superior a 8 de cada imatge.

```python
# funció que compta els píxels negres de la imatge(vector) donada
def comptarNegres(img):
    cont = 0
    for p in img:
        if p > 8:
            cont += 1
    return cont

    # aquesta linia fa el mateix
    # return np.sum(np.where(img > 8, 1, 0))
```

```python
newX_test = np.empty((0,65))
for img in X_testB:
    imgCpy = np.copy(img)
    pxs = comptarNegres(imgCpy)
    imgCpy = np.append(imgCpy, pxs)
    newX_test = np.vstack((newX_test, imgCpy))

newX_train = np.empty((0,65))
for img in X_trainB:
    imgCpy = np.copy(img)
    pxs = comptarNegres(imgCpy)
    imgCpy = np.append(imgCpy, pxs)
    newX_train = np.vstack((newX_train, imgCpy))
```

### Quantitat de cercles

Aquí es compta la quantitat de cercles que s'identifiquin a cada imatge, tal com s'ha indicat a la introducció.

```python
# funcions auxiliars per saber si en certa direcció té límit o no
def limitAmunt(img, pos):
    y = pos[0]-1
    if y < 0:
        return -1
    if img[y,pos[1]] == 0:
        return -1
    return y

def limitDreta(img, pos):
    x = pos[1]+1
    while x < 8:
        if img[pos[0],x] == 1:
            return x
        x += 1
    return -1

def limitEsquerra(img, pos) -> int:
    x = pos[1]-1
    while x >= 0:
        if img[pos[0],x] == 1:
            return x
        x -= 1
    return -1

def limitAbaix(img, pos):
    y = pos[0]
    mas_izq = 7
    mas_der = 0
    while y < 8:
        if img[y,pos[1]] == 1:
            return (mas_izq, y, mas_der)
        posI = limitEsquerra(img, (y,pos[1]))
        if posI == -1:
            return (-1, y, mas_der)
        if posI < mas_izq:
            mas_izq = posI
        posO: int = limitDreta(img, (y,pos[1]))
        if posO == -1:
            return (mas_izq, y, -1)
        if posO > mas_der:
            mas_der = posO
        y += 1
    return (mas_izq, -1, mas_der)
```

```python
# funció auxiliar per comprobar si la posició indicada a una imatge és el "començament" d'un cercle
def esCercle(img, start, end):
    limites = [7,7,0,0]
    for x in range(start[1]+1, end[1]):
        pos1 = limitAmunt(img, (start[0],x))
        if pos1 == -1:
            return [0,0,0,0]
        limites[0] = pos1
        pos2 = limitAbaix(img, (start[0],x))
        if pos2[1] == -1 or pos2[0] == -1 or pos2[2] == -1:
            return [0,0,0,0]
        if pos2[0] < limites[1]:
            limites[1] = pos2[0]
        if pos2[1] > limites[2]:
            limites[2] = pos2[1]
        if pos2[2] > limites[3]:
            limites[3] = pos2[2]

    return limites
```

```python
# funció que retorna la quantitat de cercles detectats a l'imatge donada
def comptarCercles(img):
    im = img.copy()
    im = im[:-1]
    im = np.where(np.reshape(im, (8, 8)) > 7, 1, 0)
    contador = 0
    
    for r in range(8):
        last = 0
        start = (-1,-1)
        end = (-2,-2)
        for c in range(8):
            if im[r][c] == 1:
                if last == 0:
                    if end == (-1,-1):
                        end = (r,c)
                        limites = esCercle(im, start, end)
                        if limites != [0,0,0,0]:
                            contador += 1
                        end = (-2,-2)
                start = (r,c)
            else:
                if last == 1:
                    end = (-1,-1)

            last = im[r][c]

    return contador
```

```python
newX_test2 = np.empty((0,66))
for img in newX_test:
    imgCpy = np.copy(img)
    pxs = comptarCercles(imgCpy)
    imgCpy = np.append(imgCpy, pxs)
    newX_test2 = np.vstack((newX_test2, imgCpy))

newX_train2 = np.empty((0,66))
for img in newX_train:
    imgCpy = np.copy(img)
    pxs = comptarCercles(imgCpy)
    imgCpy = np.append(imgCpy, pxs)
    newX_train2 = np.vstack((newX_train2, imgCpy))
```

### Densitat en els quadrants

Aquí es classifica de l'1 al 4 el quadrant de la imatge amb major densitat de píxels.

```python
# funció auxiliar que suma la densitat del quadrant indicat
def contarCuadrante(img, cuad, size):
    cont = 0
    for y in range(cuad[0]*size[0],size[0]+cuad[0]*size[0]):
        for x in range(cuad[1]*size[1],size[1]+cuad[1]*size[1]):
            cont += img[y][x]
    return cont
```

```python
# funció que retorna el valor assignat després de classificar l'imatge donada
def clasificarDensidad(img):
    im = img.copy()
    im = im[:-2]
    im = np.reshape(im, (8, 8))
    max = contarCuadrante(im, (0,0), (4,4))
    cuadrante = 1
    
    c = contarCuadrante(im, (0,1), (4,4))
    if c > max:
        max = c
        cuadrante = 2
        
    c = contarCuadrante(im, (1,0), (4,4))
    if c > max:
        max = c
        cuadrante = 3
        
    c = contarCuadrante(im, (1,1), (4,4))
    if c > max:
        max = c
        cuadrante = 4

    return cuadrante
```

```python
newX_test3 = np.empty((0,67))
for img in newX_test2:
    imgCpy = np.copy(img)
    dens = clasificarDensidad(imgCpy)
    imgCpy = np.append(imgCpy, dens)
    newX_test3 = np.vstack((newX_test3, imgCpy))

newX_train3 = np.empty((0,67))
for img in newX_train2:
    imgCpy = np.copy(img)
    dens = clasificarDensidad(imgCpy)
    imgCpy = np.append(imgCpy, dens)
    newX_train3 = np.vstack((newX_train3, imgCpy))
```
