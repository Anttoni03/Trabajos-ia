```python
def draw_bounding_box_tensor_direct(tensor, bbox, scale=[1,1], color=(255, 0, 0), thickness=3):
    # Asegúrate de que el tensor esté en rango [0, 1] para la imagen
    tensor = tensor.clone()  # Evitar modificar el tensor original

    # Coordenadas de la bounding box
    y_top, y_bottom, x_left, x_right = bbox
    r, g, b = color  # Descompone el color en canales

    x_left *= scale[0]
    x_right *= scale[0]
    y_top *= scale[1]
    y_bottom *= scale[1]

    x_left = int(round(x_left))
    x_right = int(round(x_right))
    y_top = int(round(y_top))
    y_bottom = int(round(y_bottom))

    # Dibuja las líneas de la bounding box en el tensor
    for t in range(thickness):
        # Líneas horizontales (superior e inferior)
        for i in range(x_left, x_right):
            tensor[:, y_top - t, i] = torch.tensor([r, g, b], dtype=tensor.dtype) / 255.0
            tensor[:, y_bottom + t, i] = torch.tensor([r, g, b], dtype=tensor.dtype) / 255.0

        for i in range(y_top, y_bottom):
        # Líneas verticales (izquierda y derecha)
            tensor[:, i, x_left - t] = torch.tensor([r, g, b], dtype=tensor.dtype) / 255.0
            tensor[:, i, x_right + t] = torch.tensor([r, g, b], dtype=tensor.dtype) / 255.0

    return tensor
```
